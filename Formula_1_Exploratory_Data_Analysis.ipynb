#**NAME OF THE DATASET: FORMULA 1 DATASET**

**Formula 1 racing originated during the 1920-30s in Europe from other similar racing competitions. In 1946, the FIA standardized racing rules and this formed the basis of Formula One racing. The inaugural Formula One World Drivers’ championship was then held in 1950, the first world championship series.** 

**Formula 1 also known as F1 is a Motorsport defined by open build single seat race cars it features the best drivers and most powerful and technically advanced cars sanctioned by the Fédération Internationale de l'Automobile also known as FIA. Here 10 team comprising of 2 drivers each compete in a period of one year to find out ultimate driver and a constructor champion. A Formula One season consists of a series of races, known as Grand Prix which take place worldwide on purpose-built circuits and on public roads.**

**The data set includes Race wins, all Race Results, Qualifying Times, Constructors as well as Drivers Championship and Fastest Laps and others for years 1950-2020.**

#**Let's merge different datasets for better understanding.**

```python
# Let's first merge results and races column.

merged_results_races = pd.merge (results, races, on = 'raceId', how = 'left')
```

```python
# Let's view what are the columns that are present in the merged dataset.

merged_results_races.columns
```

```python
# Let's view top 5 row of the dataset.

merged_results_races.head()
```

```python
# Also we can merge driver dataset with merged_results_races dataset.

merged_results_races_driver = pd.merge (merged_results_races,drivers, on = 'driverId', how = 'left')
```

```python
# Let's view the merged dataset.

merged_results_races_driver.head()
```

```python
# Let's view what are the columns that are present in the merged dataset.

merged_results_races_driver.columns
```

```python
# Again merge the merged_results_races_driver dataset with the constructor dataset.

merged_results_races_driver_constructor = pd.merge (merged_results_races_driver, constructors, on = 'constructorId', how = 'left')
```

```python
# Let's view the merged dataset.

merged_results_races_driver_constructor.head()
```

```python
# Let's view what are the columns that are present in the merged dataset.


merged_results_races_driver_constructor.columns
```

```python
# Finally merge the merged_results_races_driver_constructor dataset with the status dataset.

merged_results_races_driver_constructor_status = pd.merge (merged_results_races_driver_constructor, status, on = 'statusId', how = 'left')
```
