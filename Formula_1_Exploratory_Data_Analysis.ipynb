#**NAME OF THE DATASET: FORMULA 1 DATASET**

**Formula 1 racing originated during the 1920-30s in Europe from other similar racing competitions. In 1946, the FIA standardized racing rules and this formed the basis of Formula One racing. The inaugural Formula One World Drivers’ championship was then held in 1950, the first world championship series.** 

**Formula 1 also known as F1 is a Motorsport defined by open build single seat race cars it features the best drivers and most powerful and technically advanced cars sanctioned by the Fédération Internationale de l'Automobile also known as FIA. Here 10 team comprising of 2 drivers each compete in a period of one year to find out ultimate driver and a constructor champion. A Formula One season consists of a series of races, known as Grand Prix which take place worldwide on purpose-built circuits and on public roads.**

**The data set includes Race wins, all Race Results, Qualifying Times, Constructors as well as Drivers Championship and Fastest Laps and others for years 1950-2020.**

```python
# Let's drop the unnecessary columns.


f1df.drop(['position', 'positionText', 'time_x', 'time_y', 'url_x', 'fp1_date', 'fp1_time', 'fp2_date', 'fp2_time', 'fp3_date', 'fp3_time', 'quali_date', 'quali_time',
           'sprint_date', 'sprint_time', 'url_y', 'url'], inplace = True, axis = 1)
```

```python
# Now we can also rename the dataset for the better understanding

f1df.rename(columns = {'number_x':'number','milliseconds':'timetaken_inms','fastestLapSpeed':'maximum_speed',
 'name_x':'grand_prix','number_y':'driver_number','code':'driver_code','nationality_x':'nationality','name_y':'company', 'rank': 'fastest_lap_rank'
 }, inplace = True)
```

```python
# Let's check the column names of the dataset

f1df.columns
```

```python
f1df.info()
```

```python
# Let's replace the 'n' values with np.nan

f1df['number'].replace('\\N', np.nan, inplace = True)
f1df['timetaken_inms'].replace('\\N', np.nan, inplace = True)
f1df['fastest_lap_rank'].replace('\\N', np.nan, inplace = True)
f1df['fastestLap'].replace('\\N', np.nan, inplace = True)
f1df['maximum_speed'].replace('\\N', np.nan, inplace = True)
```

```python
# Let's change the data type of number, timetaken_inms, fastest_lap_rank, fastestLap, maximum_speed, date, dob   column to float

f1df['number'] = f1df['number'].astype(float)
f1df['timetaken_inms'] = f1df['timetaken_inms'].astype(float)
f1df['fastest_lap_rank'] = f1df['fastest_lap_rank'].astype(float)
f1df['fastestLap'] = f1df['fastestLap'].astype(float)
f1df['maximum_speed'] = f1df['maximum_speed'].astype(float)
f1df['date'] = pd.to_datetime(f1df['date'])
f1df['dob'] = pd.to_datetime(f1df['dob'])
```

```python
# Let's check the dataset 

f1df.head()
```

```python
# Let's check the columns of the dataset

f1df.columns
```

```python
# Let's add columns forename & surname to create a single column 'driver_name'

f1df['driver_name'] = f1df['forename'] + ' ' + f1df['surname']
```

```python
# Now Let's Drop The Unnecessary column of names

f1df.drop(['forename','surname'], inplace = True, axis = 1)
```

```python
# Let's see the changed dataset

f1df.head()
```

```python
# We can also check for the age of the drivers, as date of birth of the drivers are given

from datetime import datetime

f1df['driver_age'] = datetime.today()-f1df['dob']


```

```python
# Let's check the driver_age column
 
f1df['driver_age'].reset_index()
```

```python
# Let's get the years from the days

f1df['driver_age'] = round(f1df['driver_age'].dt.days/365,2)
```

```python
# We can now drop the 'dob' column, as it is not necessary.

f1df.drop(['dob'], inplace = True, axis = 1)
```

```python
# Let's check for the new

f1df.head()
```

```python
#  Let's check the dataset for the name of the drivers who own the Grand Prix maximum number of times.

drivers_with_most_wins = f1df[f1df['positionOrder'] == 1]
```

```python
# let's create a dataframe and look for the most number of wins by drivers.

drivers_with_most_wins = drivers_with_most_wins.groupby('driver_name')['positionOrder'].count().sort_values(ascending = False).to_frame().reset_index()
```

```python
# Let's see the name of the top 10 drivers with most number of wins.

drivers_with_most_wins.head(10)
```

```python
# Let's plot a barplot for the Most Number of GP Wins V/S Driver Names.

plt.figure(figsize=(10,5))
sns.set(style="darkgrid")
sns.barplot(x = "driver_name", y = 'positionOrder', data = drivers_with_most_wins.head(10))
plt.title("Drivers With Most Number of Wins", fontsize = 13)
plt.xlabel("Driver Names", fontsize = 15)
plt.ylabel("Most number of wins--> ", fontsize = 15)
plt.xticks(rotation=90)
plt.show()
```

* **It can be stated from above that Lewis Hamilton tops the rank for the most number of wins in Grand Prix match.** 

```python
#  Let's check the dataset for the name of the best constructors

constructors_with_most_wins = f1df[f1df['positionOrder'] == 1]
```

```python
# let's create a dataframe and look for the Top Constructors.

constructors_with_most_wins = constructors_with_most_wins.groupby('company')['positionOrder'].count().sort_values(ascending = False).to_frame().reset_index()
```

```python
# Let's view top 10 constructors

constructors_with_most_wins.head(10)
```

```python
# Let's plot a barplot for the Most Number of GP Wins V/S Constructor Names.

plt.figure(figsize=(12,7))
sns.set(style="darkgrid")
sns.barplot(y = "company", x = 'positionOrder', data = constructors_with_most_wins.head(10), edgecolor = 'b')
plt.title("Constructor With Most Number of Wins", fontsize = 13)
plt.ylabel("Constructor Names", fontsize = 15)
plt.xlabel("Most number of wins--> ", fontsize = 15)
plt.xticks(rotation=90)
plt.show()
```

* **It can be stated from above that Ferrari, Mclaren tops the rank for the most number of wins in Grand Prix match.**

```python
# Let's try to find the speed distribution among all the grandprix

f1df_speed = f1df[f1df['year']> 2000]
f1_speed = f1df_speed.groupby(['grand_prix','year'])['maximum_speed'].mean().to_frame().reset_index()
```

```python
# Let's see the top speeds by comparing all the grandprix

f1_speed.nlargest(20,'maximum_speed')
```

* **We can say that Italian GrandPrix has the highest average speed almost in all the year.**

```python
# Let's see the lowest speeds by comparing all the grandprix

f1_speed.nsmallest(20,'maximum_speed')
```

* **We can say that Monaco GrandPrix has the lowest average speed almost in all the year.**

```python
# Let's visualize the speed distribution among all the grandprix

speed = sns.FacetGrid(data = f1_speed, col = 'grand_prix', col_wrap = 5);
speed.map (plt.scatter, 'year', 'maximum_speed', alpha = 0.8, edgecolor = 'r', linewidth = 0.6, s = 100);
speed.set_titles("{col_name}")
speed.set_xlabels('year')
speed.set_ylabels("Average Maximum Speed")
plt.subplots_adjust(top = 0.92)
speed.fig.suptitle('Average speed amongst all teams during fastest lap in individual GPs')
plt.show();

```

* **From above we can infer that Italian Grandprix has the higest average speed and Monaco Grandprix has the lowest speed.**
